# Reading a .csv file from a URL with Pandas

### Target → football-data.co.uk/data.php

```python
import pandas as pd

# link = "football-data.co.uk/data.php"

df_sample = pd.read_csv("https://football-data.co.uk/mmz4281/2425/E0.csv")

display(df_sample)
```

### Rename columns

```python
df_sample.rename(columns={'FTHG': 'HomeGoals', 'FTAG': 'AwayGoals', 'FTR': 'Result'}, inplace=True)

display(df_sample)
```
