```python
import csv
import random


first_names = ["John", "Emily", "Michael", "Sarah", "William", "Olivia", "James", "Ava", "Robert", "Isabella"]
last_names = ["Smith", "Johnson", "Williams", "Jones", "Brown", "Davis", "Miller", "Wilson", "Moore", "Taylor"]

def generate_phone_number():
    return f"{random.randint(100, 999):03}-{random.randint(100, 999):03}-{random.randint(1000, 9999):04}"

data = []

for _ in range(1000):
    first_name = random.choice(first_names)
    last_name = random.choice(last_names)
    phone_number = generate_phone_number()
    data.append([f"{first_name} {last_name}", phone_number])

with open("random_names_and_numbers.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Name", "Phone Number"])  # Header
    writer.writerows(data)
```

```python
import pandas as pd

df = pd.read_csv("random_names_and_numbers.csv")
display(df)
```

```python
#type(df)
import pyarrow.parquet as pq

df.to_parquet("random_names_and_numbers.parquet", engine="pyarrow")
df.to_excel("random_names_and_numbers.xlsx", engine="openpyxl")

df.info()
display(df)
```
