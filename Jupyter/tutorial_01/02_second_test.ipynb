```python
# convert json to a numpy array with name and price of crypto

import numpy as np
import json
with open('../../Pandas/crypto.json') as f:
    data = json.load(f)
    data = np.array(data['data'][0]['screen_data']['crypto_data'])
    display(data)
```

```python
name_price = []
for crypto in data:
    name_price.append(crypto['name'] + ' → USD ' + crypto['inst_price_usd'])

display(name_price)
```
