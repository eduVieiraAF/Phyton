```python
import pandas as pd

#* read data.csv
data = pd.read_csv('data.csv')
display(data)
```

```python

#* extract Calories
calories = data['Calories']
display(calories)
```

```python
import matplotlib.pyplot as plt

#* plot data in a scatter chart
data.plot(kind='scatter', x='Duration', y='Calories')
plt.show()
```

```python

#* calculate mean of Calories
mean_calories = data['Calories'].mean()
display(mean_calories)
```

```python

#* calculate variance of Calories
varicance_calories = data['Calories'].var()
display(varicance_calories)
```

```python

#* calculate standard deviation of Calories
standard_deviation_calories = data['Calories'].std()
display(standard_deviation_calories)
```

```python

# Collect the highest and lowest values of Calories
highest_calories = data['Calories'].max()
lowest_calories = data['Calories'].min()

# Create a barh graph
plt.barh(['Highest', 'Lowest'], [highest_calories, lowest_calories])

# Add labels  with the values to the bar graph
plt.xlabel('Calories')
plt.ylabel('')

# Display the bar graph
plt.show()



```
